## Data reading
Body mass index (BMI) is a measure that uses your height and weight to work out if your weight is healthy.

```python
def readcsv(path):
    order = pd.read_csv(path)
    return order

order = readcsv(path)
order
```

```python
df = order[['age_group','RIDAGEYR', 'BMXBMI']]
df = pd.DataFrame(df)
df
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
```

```python
plt.figure(figsize=(13,7))
x = ['Under Weight\n<18.5', 'Normal\n18.5 - 25','Over Weight\n25 - 30','Obese\n(30 - 35)','Extreme Obese\n35>']
y = f1

bar = plt.bar(x,y,)
plt.title('Bar chart show the count of each part')
addlabels(x,y)

plt.savefig('D:/not sure but it is the Fucking BUGggg!!/MAS PROJECT/data_png/%s(BMI with count).png'%tittle)
```
